# **Performing HTTP requests in Python using urllib**

Now that you know the basics behind HTTP GET requests, it's time to perform some of your own. In this exercise, you will ping our very own DataCamp servers to perform a GET request to extract information from the first coding exercise of this course, `"https://en.wikipedia.org/wiki/Abebe_Bikila#In_popular_culture"`.

In the next exercise, you'll extract the HTML itself. Right now, however, you are going to package and send the request and then catch the response.

**Instructions:**
* Import the functions `urlopen` and `Request` from the subpackage `urllib.request.`
* Package the request to the url `"https://en.wikipedia.org/wiki/Abebe_Bikila#In_popular_culture"` using the function `Request()` and assign it to `request`.
* Send the `request` and catch the response in the variable response with the function `urlopen()`.
* Extract the response using the read() method and store the result in the variable html. Print the string html.
* Run the rest of the code to see the datatype of response and to close the connection!


````python
# Import packages
from urllib.request import urlopen, Request

# Specify the url
url = "https://en.wikipedia.org/wiki/Abebe_Bikila#In_popular_culture"

# This packages the request: request
request = Request(url)

# Sends the request and catches the response: response
response = urlopen(request)

# Catch the html content from the response
html = response.read()

# Read and print the html content
print(html)

# Print the datatype of response
print(type(response))

# Be polite and close the response!
response.close()

````

# **Turning a webpage into data using BeautifulSoup: getting the hyperlinks**

In this exercise, you'll figure out how to extract the URLs of the hyperlinks from the BDFL's webpage. In the process, you'll become close friends with the soup method `find_all()`.

**Instructions:**
* Use the method `find_all()` to find all hyperlinks in soup, remembering that hyperlinks are defined by the HTML tag `<a>` but passed to `find_all()` without angle brackets; store the result in the variable `a_tags`.
* The variable `a_tags` is a results set: your job now is to enumerate over it, using a for loop and to print the actual URLs of the hyperlinks; to do this, for every element link in `a_tags`, you want to `print() link.get('href')`.


````python
# Import packages
import requests
from bs4 import BeautifulSoup

# Specify url
url = 'https://www.python.org/~guido/'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extract the response as HTML: html_doc
html_doc = r.text

# Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html_doc, 'html.parser')

# Print the title of Guido's webpage
print("Title of the webpage:", soup.title.string)

# Find all 'a' tags (which define hyperlinks): a_tags
a_tags = soup.find_all('a')

# Print the URLs to the shell
for tag in a_tags:
    print(tag.get('href'))


````

# **Use BeautifulSoup to scrape tables and turn into pandas DataFrame?**
Here you are going to push yourself to the next level to extract table data from HTML pages using BeautifulSoup and turn it to pandas DataFrame. Doing this task requires html mark-up language and I recommend you to go through [the w3swchools html tutorials](https://www.w3schools.com/html/). For now, HTML elements can be identified with their class or id tags. Here we are going to parse the html table from `'https://www.basketball-reference.com/leagues/NBA_2021.html'` where the table is uniquely identified using the id. The id of the table we want to extract is `'per_game-team'`. The basic logic/steps to do the parsing and converting the html table to pandas DataFrame are:
1. Get the table tag
2. From the table object, Get the Header names from `<th>` tags under the `<thead>` tag
3. iterate through the rows (`<tr>` tags) and get the `<td>` content from each row

The example code is given here.

```
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://www.basketball-reference.com/leagues/NBA_2021.html'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', {'id':'per_game-team'})

headers = [x.text for x in table.find('thead').find_all('th')]

data = []
table_body_rows = table.find('tbody').find_all('tr')
for row in table_body_rows:
    rank = [row.find('th').text]
    row_data = rank + [x.text for x in row.find_all('td')]
    data.append(row_data)

df = pd.DataFrame(data, columns=headers)
```
Based on this example, your task here is to extract corona-virus information of all countries from `'url = 'https://www.worldometers.info/coronavirus/#countries'`. The table you are going to extract has an html id value of `'main_table_countries_today'`.

**Instructions:**
* Import `BeautifulSoup` from `bs4`, `requests` and `pandas` as an alias of `pd`.
* Assign the url of the html page to `url` variable.
* Get the `response` object using the `.get()` function of `requests`.
* Create a BeautifulSoup object `soup` using `BeautifulSoup()` function and passing the `text` attribute of `soup` and the `'html.parser'` attributes.
* Create an object `table` using the `.find()` function of `soup`. Please use the html tag type `'table'` and a dictionary object containing `{'id':'main_table_countries_today'}` as parameters for the `.find()` function.
* Set the DataFrame columns as `headers' list from the `table' object.
* Create an empty list `data` and assign `table_body_rows` from the `.findall()` function.
* Iterate over `table_body_rows` and append each row data into the list `data`.
* Finally create the DataFrame `df` from the `data` list and the `headers` list.
* Run the codecell to print the first five rows the `df`.



````python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Assign the URL of the HTML page
url = 'https://www.worldometers.info/coronavirus/#countries'

# Get the response object
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table with the specified id
table = soup.find('table', {'id': 'main_table_countries_today'})

# Extract the table headers
headers = table.find_all('th')

# Initialize an empty list to hold the table data
data = []

# Extract the table rows from the body
table_body_rows = table.find('tbody').find_all('tr')

# Iterate over each row and extract the cell data
for row in table_body_rows:
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells]
    data.append(row_data)

# Create a DataFrame from the data and headers
df = pd.DataFrame(data, columns=headers)

# Print the first five rows of the DataFrame
print(df.head())







````

You are now tasked to use the same approach to read data from the link, `'https://www.worldometers.info/coronavirus/#countries'` using the new short method.

**Instructions:**
* Import `pandas` as an alias of `pd`.
* Assign `url` variable to the link shown above.
* Read the DataFrame `df` from the link. Use `attr` parameter as `{'id': 'main_table_countries_today'}`.
* Print the first few rows of `df` using `.head()` function.

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Assign the URL
url = 'https://www.worldometers.info/coronavirus/#countries'

# Fetch the webpage
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table by its ID
table = soup.find('table', {'id': 'main_table_countries_today'})

# Read the HTML table into a DataFrame
try:
    df = pd.read_html(url, attrs={'id': 'main_table_countries_today'})[0]
    # Print the first few rows of the DataFrame
    print(df.head())
except ValueError as e:
    print(f"Error: {e}. The table might be dynamically loaded.")

# Print the first few rows
print(df.head())
````
