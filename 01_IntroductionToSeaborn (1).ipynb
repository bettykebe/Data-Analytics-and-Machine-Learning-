# **Making a scatter plot with lists**

In this exercise, we'll use a dataset that contains information about 227 countries. This dataset has lots of interesting information on each country, such as the country's birth rates, death rates, and its gross domestic product (GDP). GDP is the value of all the goods and services produced in a year, expressed as dollars per person.

We've created three lists of data from this dataset to get you started. `gdp` is a list that contains the value of GDP per country, expressed as dollars per person. `phones` is a list of the number of mobile phones per `1,000` people in that country. Finally, `percent_literate` is a list that contains the percent of each country's population that can read and write. These three variables are created in the first few lines of code in the codecell given below for you to use.

**Instructions:**
* Create a scatter plot of GDP (`gdp`) vs. number of phones per 1000 people (`phones`).
* Display the plot.
* Change the scatter plot so it displays the percent of the population that can read and write (`percent_literate`) on the y-axis.


```python
#Loading the DataFrame from the dataset file
countries =  pd.read_csv('https://github.com/DataAnalyst21/DatasetsForDataAnalytics/blob/main/countries-of-the-world.csv?raw=True')

#Displaying the first 5 rows of the DataFrame
print(countries.head())

#Creating the gdp, phones and percent_literate lists from countries DataFrame
gdp = list(countries['GDP ($ per capita)'])
phones = list(countries['Phones (per 1000)'])
percent_literate = list(countries['Literacy (%)'])

#Exploring the first few elements by displaying the elements using print
print(gdp[:5])
print(phones[:5])
print(percent_literate[:5])


# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(data=countries, x="GDP ($ per capita)", y="Phones (per 1000)")
plt.title("GDP vs. Number of Phones")
plt.show()

# Create scatter plot with GDP on the x-axis and percent literate on the y-axis
sns.scatterplot(data=countries, x="GDP ($ per capita)", y="Literacy (%)")
plt.title("GDP vs. Percent Literate")
plt.show()



```

# **Making a count plot with a list**

In the last exercise, we explored a dataset that contains information about 227 countries. Let's do more exploration of this data - specifically, how many countries are in each region of the world?

To do this, we'll need to use a count plot. Count plots take in a categorical list and return bars that represent the number of list entries per category. You can create one here using a list of regions for each country, which is a variable named `region`. The `region` list is created in the first few lines of code below from `countries` DataFrame.

**Instructions:**
* Use Seaborn to create a count plot with `region` on the y-axis.
* Display the plot.


```python
#Creating the region list from countries DataFrame
region = list(countries['Region'])


#Exploring the first few elements by displaying the elements using print


print(region[:5])



# Create count plot with region on the y-axis
sns.countplot(y=region, )


# Adding a title and labels
plt.title("Number of Countries by Region")
plt.xlabel("Number of Countries")
plt.ylabel("Region")

# Displaying the plot
plt.tight_layout()

plt.show()


```

# **Hue and scatter plots**

We have learned how hue allows us to easily make subgroups within Seaborn plots in our class. Let's try it out by exploring data from students in secondary school. We have a lot of information about each student like their age, where they live, their study habits and their extracurricular activities.

For now, we'll look at the relationship between the number of absences they have in school and their final grade in the course, segmented by where the student lives (rural vs. urban area).

The `student_data`DataFrame is loaded for you in the first few lines of code in the codecell below.

**Instructions:**
* Create a scatter plot with `"absences"` on the x-axis and final grade (`"G3"`) on the y-axis using the DataFrame `student_data`. Color the plot points based on `"location"` (urban vs. rural).
* Make `"Rural"` appear before `"Urban"` in the plot legend.




```python
#Loading students_data DataFrame from dataset file
student_data = pd.read_csv('https://github.com/DataAnalyst21/DatasetsForDataAnalytics/blob/main/student-alcohol-consumption.csv?raw=True')

#Displaying the first 5 rows of the DataFrame
print(student_data.head())

# Create a scatter plot of absences vs. final grade

sns.scatterplot(x="absences", y="G3", hue="location", data=student_data, hue_order=["Rural", "Urban"])

# Show plot
plt.show()



```

# **Hue and count plots**

Let's continue exploring our dataset from students in secondary school by looking at a new variable. The `"school"` column indicates the initials of which school the student attended - either `"GP"` or `"MS"`.

In the last exercise, we created a scatter plot where the plot points were colored based on whether the student lived in an urban or rural area. How many students live in urban vs. rural areas, and does this vary based on what school the student attends? Let's make a count plot with subgroups to find out.

**Instructions:**
* Fill in the `palette_colors` dictionary to map the `"Rural"` location value to the color `"green"` and the `"Urban"` location value to the color `"blue"`.
* Create a count plot with `"school"` on the x-axis using the `student_data` DataFrame.
* Add subgroups to the plot using `"location"` variable and use the `palette_colors` dictionary to make the location subgroups green and blue.


```python

# Create a dictionary mapping subgroup values to colors
import seaborn as sns
import matplotlib.pyplot as plt

# Define the color palette
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot with 'school' on the x-axis and subgroups by 'location'
sns.countplot(x="school", hue="location", data=student_data, palette=palette_colors)




# Display plot
plt.show()
```
